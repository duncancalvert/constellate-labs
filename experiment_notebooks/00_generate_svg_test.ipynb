## Simple Generate SVG Tester

```python
from pathlib import Path

from constellate_labs.pipeline import generate_svg

# User prompt: describe the shape or object you want as SVG
prompt = "a simple smiley face"

# Generate SVG (uses fallback circle when llm_call=None; pass a callable for real LLM)
result = generate_svg(prompt, canvas_width=100, canvas_height=100)

# Export SVG to a file
output_path = Path("output.svg")
output_path.write_text(result.svg_content, encoding="utf-8")

print(f"Prompt: {prompt}")
print(f"Wrote {len(result.svg_content)} chars to {output_path.resolve()}")
```

```python
# Optional: display the SVG in the notebook (if running in Jupyter/IPython)
try:
    from IPython.display import SVG, display
    display(SVG(result.svg_content))
except ImportError:
    print("(Install ipython to display SVG inline, or open output.svg in a browser.)")
```
